```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
import statistics as stats
```

```python
#Cargar el archivo en una variable llamada autos
autos=pd.read_csv("Autos.csv", encoding="ISO-8859-1")
```

```python
#Mostrar datos estadisticos de los precios
print("Estadisticas descriptivas:")
autos.price.describe()
```

```python
autos.head()
```

```python
autos.info()
```

```python
#Mostrar valor minimo de los datos sobre los precios
print("MINIMO:")
autos.price.min()
```

```python
#Mostrar el valor maximo de los datos sobre los precios
print("MAXIMO:")
autos.price.max()
```

```python
#Calcular la moda de los datos sobre los precios
print("MODA:")
moda=stats.mode(autos.price)
print(moda)
```

```python
#Calcular la mediana de los datos sobre los precios
print("MEDIANA:")
mediana=autos.price.median()
autos.price.median()
```

```python
#Calcular la media de los datos sobre los precios
print("MEDIA:")
media=autos.price.mean()
autos.price.mean()
```

```python
#Calcular la desviacion estandar de los datos sobre los precios
print("DESVIACION ESTANDAR:")
desviacion=autos.price.std()
autos.price.std()
```

```python
#Mostrar un histograma
plt.hist(autos.price, alpha=0.2, color='b', edgecolor="black")
plt.axvline(moda, color='r', linestyle='dashed', linewidth=2, label=f'Moda: {moda:.2f}')
plt.axvline(mediana, color='b', linestyle='dashed', linewidth=2, label=f'Mediana: {mediana:.2f}')
plt.axvline(media, color='y', linestyle='dashed', linewidth=2, label=f'Media: {media:.2f}')
plt.axvline(media - desviacion, color='green', linestyle='dashed', linewidth=2, label=f'Desviación: {desviacion:.2f}')
plt.axvline(media + desviacion,color='green',linestyle='dashed',linewidth=2)
plt.legend()
plt.title("Histograma de los precios de los autos")
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
```

##**CONCLUSIONES:**
##Nos podemos dar cuenta que normalmente los autos de menor precio son adquiridos con mas frecuencia siendo desde 5.000 hasta 9,000 los mas frecuentes.
##Tambien es interesente observar que a partir de los 20,000 empieza a bajar la frecuencia pero aparece un pico entre 34,000 hasta $37,000 y después la frecuencia vuelve a bajar mas.

```python
drive_counts = autos['body-style'].value_counts()
# Mostrar gráfico de barras
plt.figure(figsize=(10,5))
plt.bar(drive_counts.index, drive_counts.values, color='b', alpha=0.6, linestyle='dashed')
plt.title("Distribución de los tipos de vehiculos (Body Style)")
plt.xlabel("Tipo de Vehiculo")
plt.ylabel("Precio")
```

```python
drive_counts = autos['drive-wheels'].value_counts()
#Mostrar gráfico de barras
plt.figure(figsize=(10,7))
plt.bar(drive_counts.index, drive_counts.values, color='b', alpha=0.6, linestyle='dashed')
plt.title("Distribución de los tipos de tracción (Drive Wheels)")
plt.xlabel("Tipo de tracción")
plt.ylabel("Precio")
```

```python
drive_counts = autos['engine-location'].value_counts()
#Mostrar gráfico de barras
plt.figure(figsize=(8,7))
plt.bar(drive_counts.index, drive_counts.values, color='b', alpha=0.5, linestyle='dashed')
plt.title("Distribución de la ubicacion del motor (Engine Location)")
plt.xlabel("Ubicacion del motor")
plt.ylabel("Precio")
```

```python
#Diagrama de Caja y de BIgotes
sns.boxplot(x="body-style", y="price", data=autos)
plt.title(" Tipo de vehiculo vs Precio")
plt.xlabel("Tipo de vehículo")
plt.ylabel("Precio")
```

##**CONCLUSIONES**
##Con los diagramas se puede observar de cuanto es la tendencia, mientras que unos buscan un sedan otra parte ve el convertible mas barato.
##Aunque en los diagramas de cajas se ve mas especifico como es la cantidad promedio y sus extremos en cada uno de los vehiculos.

```python
sns.boxplot(x="drive-wheels", y="price", data=autos)
plt.title("Tipo de traccion vs Precio")
plt.xlabel("Tipo de traccion")
plt.ylabel("Precio")
```

##**CONCLUSIONES**
##Podemos observar que el tipo de traccion rwd tiene un precio mas alto en promedio de los demas, inclusive teniendo casos donde se extiende mas del doble del promedio, aunque son casos limitados.

```python
sns.boxplot(x="engine-location", y="price", data=autos)
plt.title("Ubicacion del motor vs Precio")
plt.xlabel("Ubicacion del motor")
plt.ylabel("Precio")
```

##**CONCLUSIONES**
