```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
from matplotlib import pyplot
from statistics import *
```

```python
#cargar el archivo CSV en una variable llamada covid
covid = pd.read_csv("DatosCOVID.csv", encoding= "ISO-8859-1")
```

```python
covid.head()
```

```python
covid.info()
```

```python
#Mostrar la edad maxima de la base de datos de la columna edad
covid.EDAD.max()
```

```python
#Mostrar la edad minima de la base de datos de la columna edad
covid.EDAD.min()
```

```python
#Mostrar datos estadisticos de la columna edad
covid.EDAD.describe()
```

```python
#Crear el histograma
media=covid.EDAD.mean()
plt.pyplot.hist(covid.EDAD, bins=11)
plt.pyplot.title("Histograma de edades Pruebas COVID")
plt.pyplot.axvline(media, color='y', linestyle='dashed', linewidth=2, label=f'Media={media:.2f}')
plt.pyplot.xlabel("Edades")
plt.pyplot.ylabel("Frecuencia")
```

```python
#Mostrar un histograma de la columna de sexo para comparar cantidad de cada grupo
plt.pyplot.hist(covid.SEXO, bins=3)
plt.pyplot.title("Hisograma Sexo de Pruebas COVID")
plt.pyplot.xlabel("Sexo")
plt.pyplot.ylabel("Frecuencia")
```

```python
plt.pyplot.hist(covid.TIPO_PACIENTE, bins=3)
plt.pyplot.title("Hisograma Tipo Paciente de las pruebas COVID")
plt.pyplot.xlabel("Tipo Paciente")
plt.pyplot.ylabel("Frecuencia")
```
