```python
import random
```

```python
# Creamos una lista vacia que representa la edad de 100 personas
edades=[]

#Hacemos un ciclo para generar 100 numeros aleatorios que representan la edad de las personas
for i in range(100):
  # generamos un numero aleatorio con la funcion randint y lo guardamos en la variable numero_aleatorio
  numero_aleatorio = random.randint( 18, 70)
  #Agregamos el numero aleatorio a la lista de valores
  edades.append(numero_aleatorio)

print(edades)
```

```python
import pandas as pd
```

```python
#Crear un DataFrame de pandas con los numeros aleatorios
df= pd.DataFrame(edades,columns=['Edad'])
df.head() #
```

```python
#Calcular la tabla de distribucion de frecuencias
tabla_frecuencias= df['Edad'].value_counts().reset_index()
tabla_frecuencias.columns=['Edad','Frecuencia']
tabla_frecuencias= tabla_frecuencias.sort_values(by='Edad')

#Imprimir la tabla de distribucion de frecuencias
print("Tabla de Distribucion de Frecuencias: ")
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Relativa'] = tabla_frecuencias['Frecuencia'] / 100
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Porcentual'] = tabla_frecuencias['Frecuencia'] / 100*100
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Acumulada'] = tabla_frecuencias['Frecuencia'].cumsum() #Cumsum para calcular la suma acomulada
tabla_frecuencias
```

#Regla de Struges
1.Es una regla que sirve para calcular el numero de clases o intervalos idoneo en los que se debe dividir un conjunto de datos.
2.Establece que un numero de clases es igual a uno mas el logratimos en vase dos del numero total de datos. c=1+log2(N) donde N es el tamaño de la muestra

```python
# Importamos la libreria math de funciones matematicas
import math
```

```python
# Volvemos a usar la lista de edades
print(edades)
```

```python
#Crear un DataFrame con las edades
df = pd.DataFrame({'Edad': edades})
df
```

```python
#Calcular el  numero de intervalos utilizando la regal de Sturges
n = len(edades)
K = int(1+math.log2(100))
print("Valor de N = igual ", n, " \nNumero de intervalos =", K)
```

```python
#Crear la tabla de frecuencia
tabla_frecuencia=pd.cut( df['Edad'], bins=K, include_lowest=True).value_counts().reset_index()
tabla_frecuencia.columns = ['Clase', 'Frec']
tabla_frecuencia
```

```python
#Ordenar la tabla de frecuencia
tabla_frecuencia=tabla_frecuencia.sort_values(by='Clase') #sort es para ordenar
tabla_frecuencia
```

```python
#Generar columnas siguientes
tabla_frecuencia['Frec Rel']=tabla_frecuencia['Frec']/n
tabla_frecuencia['Frec Porc']=tabla_frecuencia['Frec']/n*100
tabla_frecuencia['Frec Ac']=tabla_frecuencia['Frec'].cumsum()
tabla_frecuencia['Frec Rel Ac']=tabla_frecuencia['Frec Rel'].cumsum()
tabla_frecuencia['Frec Porc Ac']=tabla_frecuencia['Frec Porc'].cumsum()

#Mostrar la tabla de frecuencia
print(tabla_frecuencia)
```
